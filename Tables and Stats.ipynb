```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", 30)
```

```python
trn = pd.read_csv('A:/shared/P428/Datasets/train_data.csv')
tst = pd.read_csv('A:/shared/P428/Datasets/test_data.csv')
df = pd.concat([trn, tst]).dropna().drop(['ID', 'Unnamed: 0'], axis=1)
```

```python
trn.dropna().drop('Unnamed: 0', axis=1).to_csv('A:/shared/P428/Datasets/train_data_clean.csv', index=False)
tst.dropna().drop('Unnamed: 0', axis=1).to_csv('A:/shared/P428/Datasets/test_data_clean.csv', index=False)
```

```python
df_cat = ['Gender', 'RaceEthnicity', 'LivingSituation', 'Education', 'BMI',
          'CurrSmoking', 'AlcUse', 'Diabetes', 'Hypertension', 'Dyslipidemia',
          'Frailty', 'FamHistHA', 'HistCancer', 'RegAspirin', 'Endpoint',
          'LipidLoweringAgents', 'AntihypertensiveAgents']
df_num = [col for col in df.columns if col not in df_cat]
df_num.remove('Treatment')
```

```python
trt_grp = df.groupby('Treatment')
trt_grp.count()
for i, j in trt_grp:
    for col in df_cat:
        display(j[col].value_counts())
```

```python
from scipy.stats import ttest_ind
for col in df_num:
    cols = []
    for i, j in trt_grp:
        cols.append(j[col])
    tval, pval = ttest_ind(a=cols[0], b=cols[1], alternative='two-sided', equal_var=False)
    print(col, pval)
    
```
